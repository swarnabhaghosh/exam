```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
data = load_digits()
```

```python
X = data.data
y = data.target
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
```

```python
X_decomposed = pca.fit_transform(X)
```

```python
print("Original shape:", X.shape)
print("Reduced shape:", X_decomposed.shape)
```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
sum = 0
for i in pca.explained_variance_ratio_:
    sum += i
print("total variance kept = ", sum)
```

```python
scatter = plt.scatter(
    X_decomposed[:, 0],      # PCA Component 1 (x-axis)
    X_decomposed[:, 1],      # PCA Component 2 (y-axis)
    c = y,                   # Color each point based on its class label
    cmap="tab10",            # 10-color colormap (good for multi-class)
    alpha=0.7,               # Slight transparency
    s=40                     # Marker size
)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA on Digits Dataset (64D  2D)")
plt.colorbar(scatter , label="Digit Label")
plt.grid(True)
plt.show()
```

```python
pca_full = PCA().fit(X) # PCA with all components
plt.plot(np.cumsum(pca_full.explained_variance_ratio_), marker='o') # np.cumsum(...) : Computes cumulative sum
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Explained Variance vs Components")
plt.grid(True)
plt.show()
```
