```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd . DataFrame ({
'Hours': [1 , 2 , 3 , 4.5 , 5 , 6 , 7 , 8.5 , 9 , 10] ,
'Marks': [35 , 40 , 50 , 60 , 62 , 70 , 75 , 85 , 88 , 95]
})
```

```python
X = data[['Hours']]
y = data['Marks']
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X, y)
```

```python
y_pred = lr.predict(X)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, root_mean_squared_error, r2_score

mse = mean_squared_error(y, y_pred)
rmse = root_mean_squared_error(y, y_pred)
mae = mean_absolute_error(y, y_pred)
r2score = r2_score(y, y_pred)

print('mse =', mse)
print('rmse =', rmse)
print('mae =', mae)
print('r2score =', r2score)
```

```python
# plot

plt.scatter(X, y, color = 'blue', label = 'actual marks')
plt.plot(X, y_pred, color = 'red', label = 'Regression line')
plt.title('study hours vs marks')
plt.xlabel('Hours studied')
plt.ylabel('Marks obtained')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# prediction

test_hours = [[6.5]]
predicted_marks = lr.predict(test_hours)
print(f"Predicted marks for 6.5 study hours : {predicted_marks [0]:.2f}")
```
