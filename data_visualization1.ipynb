```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Q1

```python
x=np.array(["A", "B", "C", "D"])
y=np.array([4.9, 5.6, 5.11, 5.2])
```

```python
plt.title("Syudents name vs Heights")
plt.xlabel("Students name")
plt.ylabel("Heights")
plt.bar(x, y, width = .25) #using matplotlib
```

```python
sns.barplot(x=x, y=y) #using seaborn
plt.title("Syudents name vs Heights")
plt.xlabel("Students name")
plt.ylabel("Heights")
plt.show()
```

### Q2

```python
data={'Year of Production':['2018','2019','2020','2021'],'Production Quantity':[58000,70000,95000,69000]}
```

```python
df1=pd.DataFrame(data) #creating dataframe from dictionary
df1
```

```python
plt.bar(df1["Year of Production"], df1["Production Quantity"])
plt.xlabel("Year of Production")
plt.ylabel("Production Quantity")
plt.title("Year of Production Vs Production Quantity")
plt.show()
```

### Q4

```python
iris = sns.load_dataset("iris")
```

```python
iris.head(10)
```

```python
# count number of flowers in each category
counts = iris.species.value_counts()
counts
```

```python
# plot the counts using plt.pie
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=90)
plt.show()
```
